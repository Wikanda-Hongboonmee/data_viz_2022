## Quiz (ให้ใช้เฉพาะที่อาจารย์สอนไปแล้วในวิชานี้)

1.คำนวณ อายุเฉลี่ย ของ**ผู้ชาย** ในข้อมูลทั้งหมด

````
# 1. คำนวณอายุเฉลี่ยของผู้ชาย ในข้อมูลทั้งหมด
this_data = data_covid[['sex','age','province_of_onset']] # กำหนดดัวแปรชื่อ this_data เพื่อนำมาใช้ในการเก็บข้อมูลคอลัมน์ของ sex,age และ province_of_onset
````

````
this_data # แสดงตารางข้อมูลของ this_data
````

````
male = this_data[this_data['sex']=='ชาย'] # เลือกเฉพาะเพศชาย กำหนดให้ 'sex' == 'ชาย' โดยเลือกจากตาราง this_data แล้วเอาไปใส่ในตารางที่ชื่อว่า male
male['age'] # แสดงตารางข้อมูลของ male คอลัมน์ 'age'
````

````
# สามารถคำนวณอายุเฉลี่ยของผู้ชาย ในข้อมูลทั้งหมด ได้ดังนี้
sum = 0 # กำหนดให้ sum = 0 เนื่องจากถ้าค่าตัวที่ 1 เข้ามาบวกจะได้ค่าออกมาเท่ากับตัวที่ 1
N = 0 # กำหนดให้ N = 0 เนื่องจาก N เป็นตัวหาร โดยจะทำการวนลูป (N เป็นตัวแปรที่จะนับว่ามีคนทั้งหมดกี่คน)
for a in male['age']: # เริ่มวนลูปโดยให้ a แทนค่าแต่ละค่าในคอลัมน์ 'age' ของตารางที่มีชื่อว่า male  
    if a > 0: # ถ้า a มากกว่า 0 แสดงว่าเป็นอายุปกติ ไม่ใช่ข้อมูลที่ผิดพลาด 
        sum += a # sum = sum + a คือ ถ้า a > 0 จะนำมารวมเข้าไปใน sum เรื่อย ๆ ในทุก ๆ รอบที่ a มากกว่า 0
        N += 1 # คำนวนจำนวนโดยที่มันจะ +1 เข้ามาเรื่อย ๆ ตามจำนวน
        # sum คือ ค่าของผลรวมของอายุทั้งหมด  N คือ ค่าของจำนวนคนทั้งหมด
print(f'อายุเฉลี่ยของผู้ป่วยชาย เท่ากับ {sum/N} ปี') # sum/N = อายุเฉลี่ยของผู้ชายในข้อมูลทั้งหมด
# จะได้ว่า อายุเฉลี่ยของผู้ป่วยชาย เท่ากับ 39.46076940331987 ปี
````

2.คำนวณ อายุเฉลี่ย ของ**ผู้ชาย**ในจังหวัดขอนแก่น

````
male_KhonKaen = male[male['province_of_onset']=='ขอนแก่น']  # เลือกเฉพาะจังหวัดขอนแก่น กำหนด 'province_of_onset' == 'ขอนแก่น' โดยเลือกจากตาราง male แล้วเอาไปใส่ในตัวแปรชื่อ male_KhonKaen
male_KhonKaen # แสดงตารางข้อมูลของ male_KhonKaen
````

````
#สามารถคำนวณอายุเฉลี่ยของผู้ชายในจังหวัดขอนแก่น ได้ดังนี้
sum = 0 # กำหนดให้ sum = 0 เนื่องจากถ้าค่าตัวที่ 1 เข้ามาบวกจะได้ค่าออกมาเท่ากับตัวที่ 1
N = 0 # กำหนดให้ N = 0 เนื่องจาก N เป็นตัวหาร โดยจะทำการวนลูป (N เป็นตัวแปรที่จะนับว่ามีคนทั้งหมดกี่คน)
for a in male_KhonKaen['age']: #เริ่มวนลูปโดยให้ a แทนค่าแต่ละค่าในคอมลัมน์ 'age' ของตารางที่มีชื่อว่า male_KhonKaen 
    if a > 0: # ถ้า a มากกว่า 0 แสดงว่าเป็นอายุปกติ ไม่ใช่ข้อมูลที่ผิดพลาด 
        sum += a # sum = sum + a คือ ถ้า a > 0 จะนำมารวมเข้าไปใน sum เรื่อย ๆ ในทุก ๆ รอบที่ a มากกว่า 0
        N += 1 # คำนวนจำนวนโดยที่มันจะ +1 เข้ามาเรื่อย ๆ ตามจำนวน
        # sum คือ ค่าของผลรวมของอายุทั้งหมด  N คือ ค่าของจำนวนคนทั้งหมด
print(f'อายุเฉลี่ยของผู้ป่วยชายในจังหวัดขอนแก่น เท่ากับ {sum/N} ปี') # sum/N = อายุเฉลี่ยของผู้ชายในจังหวัดขอนแก่น
# จะได้ว่า อายุเฉลี่ยของผู้ป่วยชายในจังหวัดขอนแก่น เท่ากับ 26.571428571428573 ปี
````

### การวนลูป record ในตาราง  (```.iterrows```)

### การวนลูป แบบมองตารางแพนด้าส์(pandas dataframe) เป็น numpy array  (```.iloc```)

````
for each_row in range(this_data.shape[0]):
    if (this_data.iloc[each_row,1] == 20) and (this_data.iloc[each_row,2] == 'ขอนแก่น'):  #each_row คือ ตัวเลข
        print(each_row)
        print(this_data.iloc[each_row,:])   
````
