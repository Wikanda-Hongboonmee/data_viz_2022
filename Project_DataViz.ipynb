# **รวมตาราง**

```python
data_covid = pd.concat([data_01, data_02, data_03, data_04])
data_covid
```

## **ตรวจสอบค่า Missing ของตาราง data_covid**

```python
data_covid.isnull().any() # ดูว่าคอลัมน์ไหนมีค่า missing บ้าง สรุปคือทุกคอลัมน์ไม่มีค่า  missing
```

### **แก้ค่า Missing ของคอลัมน์ 'age'**

```python
# แทนค่า Missing ของคอลัมน์ 'age' ด้วยค่า mean ของ 'age'
data_covid[['age']].mean()       
avg_age = data_covid[['age']].mean() 
data_covid[['age']] = data_covid[['age']].fillna(avg_age)   # แทนค่าเฉลี่ยของอายุลงไปในค่าว่าง
```

```python
data_covid.isnull().any() # ดูว่าคอลัมน์ไหนมีค่า missing บ้าง จะเห็นว่าคอลัมน์ 'age' เป็น False นั่นคือไม่มีค่า missing
```
